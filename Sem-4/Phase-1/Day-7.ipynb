```python
import networkx as nx
```

```python
g = nx.Graph()
g.add_node('A')
g.add_node('B')
g.add_node('C')
g.add_node('D')
g.add_edge('A',"B")
g.add_edge('B',"C")
g.add_edge('C',"A")
g.add_edge('A',"A")
g.add_edge('D',"B")
g.add_edge('D',"A")
nx.draw(g,with_labels=True)
```

```python
g = nx.DiGraph()
g.add_node('A')
g.add_node('B')
g.add_node('C')
g.add_node('D')
g.add_edge('A',"B")
g.add_edge('B',"C")
g.add_edge('C',"A")
g.add_edge('A',"A")
g.add_edge('D',"B")
g.add_edge('D',"A")
nx.draw(g,with_labels=True)
```

```python
g = nx.Graph()
g.add_nodes_from([1,2,3])
g.add_edges_from([(1,2),(2,3),(3,1),(1,1)])
nx.draw(g,with_labels=True)
```
