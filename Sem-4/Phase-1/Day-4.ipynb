```python
import numpy as np
import pandas as pd

df = pd.read_csv('auto-mpg.csv')
```

```python
df = df[df['horsepower']!='?']
```

```python
df['horsepower']=df['horsepower'].astype('int')
```

```python
df.dtypes
```

```python
df.corr()  #numeric_only=True if new version of pandas
```

```python
import seaborn as sls
```

```python
sls.heatmap(df.corr(),annot=True)
```

```python
sls.heatmap(df.corr(),annot=True,vmin=0)
```

```python
sls.heatmap(df.corr(),annot=True,vmax=0.5)
```

```python
sls.heatmap(df.corr(),annot=True,vmin=0,linewidth=5,linecolor="pink",cmap='coolwarm')
```

```python
sls.scatterplot(data=df,x='horsepower',y='weight',hue='cylinders',style='cylinders',s=100)
```
