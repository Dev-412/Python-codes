```python
def Grade(Total):
    if Total>200:
        return "A+"
    if Total>200:
        return "A"
    if Total>200:
        return "B"
    if Total>200:
        return "C"
    return "F"
```

```python
df["Grade"]=df["Total"].apply(Grade)
```
