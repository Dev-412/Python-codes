```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("fish.csv")
df.head()
```

```python
# for predicting weight from length

x= df[['Length1']]
y= df[['Weight']]

x,y
```

```python
# split in two part train and test

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## Build a model

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
x_train
```

```python
y_pred = model.predict(x_test)
```

```python
plt.plot(x_test,y_pred)
plt.scatter(x_test,y_test)
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```

```python
# plot mse,r2_score for test size(0.75,0.8,0.85,0.9,0.95)

df=pd.read_csv('Fish.csv')
x=df.drop(['Species','Weight'],axis=1)
y=df[['Weight']]
test=[0.75,0.8,0.85,0.9,0.95]
mse=[]
r2=[]
for i in test:
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=i,random_state=40)
    model=LinearRegression()
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    mse.append(mean_squared_error(y_test,y_pred))
    r2.append(r2_score(y_test,y_pred))
```

```python
plt.plot(test,mse)
```
