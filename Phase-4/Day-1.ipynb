```python
class a:
    def __init__(self):
        print("hello")
    
    def show(self):
        print("show")
        
class b(a):
    def haha(self):
        super().show()
        print("haha")

c = b()
c.haha()

b.__mro__
# or
b.mro()
```

```python
class student:
    def __init__(self,x):
        self.x=x
    def __add__(self,other):
        return(self.x+other.x)
    
    def __lt__(self,other):
        return("less")
    
    def __gt__(self,other):
        return("greater")
    
        
s1 = student(10)
s2 = student(20)

print(s1+s2)
print(s1>s2)
print(s1<s2)
```
