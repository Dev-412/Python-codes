```python
#Scatter

p=['t1','t2','t3','t4']
c1=['red','blue','black','green']
x =[30,20,40,50]
y =[50,10,30,60]
z =[25,30,60,30]
s1=[800,200,400,500]
plt.scatter(p,x,c=c1,s=s1,alpha=0.4,marker='*')
plt.scatter(p,y,c=c1,s=s1,alpha=0.2)
plt.scatter(p,z,c=c1,s=s1,alpha=0.7)
plt.show()
```

```python
#bar
name=["dev",'dhruv','dhoni','kohli']
runs=[72,54,77,18]
plt.bar(name,runs,color='pink',width=0.5)
plt.show()
plt.barh(name,runs,color='cyan',height=0.5)
plt.show()
```

```python
#pie
y =[30,20,40,50]
run=['t1','t2','t3','t4']
e=[0,0.1,0,0.3]
plt.pie(y,labels=run,startangle=90,explode=e,autopct='%1.1f')
plt.legend(title='Marks')
plt.show()
```
