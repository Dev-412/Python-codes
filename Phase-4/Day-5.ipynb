```python
import numpy as np
scores=[[13,10,9,33],[63,46,90,42],[39,76,13,29],[82,9,29,78],[67,61,59,36]]
x = np.array(scores)
# print(x)
match_5=np.array([41,87,72,36,92])
match_5 = match_5.reshape(5,1)
# print(match_5)
ans = np.concatenate((x,match_5),axis=1)
print(ans)
print("----------------------------------")

bat_6=np.array([77,83,98,95,89])
bat_6=bat_6.reshape(1,5)
bat_7=np.array([92,71,52,61,53])
bat_7=bat_6.reshape(1,5)

ans = np.concatenate((ans,bat_6),axis=0)
ans = np.concatenate((ans,bat_7),axis=0)
print(ans)
print("----------------------------------")

l=np.sum(ans,axis=1)
l = np.array(l)
l=l.reshape(7,1)
print(l)
ans = np.concatenate((ans,l),axis=1)
print(ans)
```
