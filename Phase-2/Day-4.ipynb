```python
y=[1,2,3,4,5,6,7]

y.sort()
print(f"max = {y[-1]} 2nd max = {y[-2]} min = {y[0]} 2nd min = {y[1]}")

p = [i for i in y if i%2==0]
print(p)
print(f"max={p[-1]} min={p[0]}")
q = [i for i in y if i%2!=0]
print(q)
print(f"max={q[-1]} min={q[0]}")
```

```python
x =["pyt","pytho","python","pyth"]
x.sort(reverse=True)
ans=""

for i in range(len(x[-1])):
    if(x[0][i]==x[-1][i]):
        ans = ans+x[-1][i]
        continue
    else:
        break
print(ans)
```
