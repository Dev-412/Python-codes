```python
# min 8 length atleast 1 up 1 lower 1 digit 1 special

x =input("Enter pass: ")

a=0
b=0
c=0
d=0
e=0
if(len(x)>=8):
    for i in x:
        if(i.isupper()):
            a=a+1
        elif(i.islower()):
            b=b+1
        elif(i.isnumeric()):
            c=c+1
        elif(i.isspace()):
            e=e+1
        else:
            d=d+1
    if(a>=1 and b>=1 and c>=1 and d>=1 and e==0):
        print("valid")
    else:
        print("invalid")
else:
    print("invalid")
```
