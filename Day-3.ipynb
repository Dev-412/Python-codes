```python
import numpy as np
z = np.arange(1,501)
print(z)
```

```python
z=z.reshape(5,4,25)
print(z)
```

```python
print(z[3,2,-1])
```

```python
# concatenate
x = [1,2,3,4,5,6,7,8,9]
y=[10,11,12,13,14,15,16,17,18]

a1 =np.array(x).reshape(3,3)
a2 =np.array(y).reshape(3,3)
print(a1)
print("----------------------------------")
print(a2)
print("----------------------------------")
z = np.concatenate((a1,a2),axis=0)
print(z)
print("----------------------------------")
z = np.concatenate((a1,a2),axis=1)
print(z)
print("----------------------------------")
```

```python
#Array Split
x = np.arange(1,11)
z = np.array(x)
print(np.array_split(z,3))
print(np.array_split(z,6))
```

```python
#where filter
x = np.arange(1,11)
z = np.array(x)
print(z)
print(np.where(z==7)) #index position
print(np.where(z%2==0)) #index position
print(np.where(z%2==0,"even","odd"))
```

```python
#Sorting
x = [[54,7,4,9],[12,46,32,75],[89,34,45,43]]
z = np.array(x)
print(z)
print("----------------------------------")
print(np.sort(z)) #row wise
print("----------------------------------")
print(np.sort(z,axis=0)) #column wise
print("----------------------------------")
print(np.sort(z,axis=1)) #row wise
```

```python
# numpy iteration
print(z)
for i in np.nditer(z):
    print(i)
```
