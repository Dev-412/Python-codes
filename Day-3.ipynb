```python
class Solution:
    def twoSum(self, nums, target):
        for i in range(len(nums)):
            for j in range(len(nums)):
                if(nums[i]==nums[j]):
                    continue
                if(nums[i]+nums[j]==target):
                    break
            
            if(nums[i]+nums[j]==target):
                break
        
        return(i,j)
```
