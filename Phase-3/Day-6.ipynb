```python
class ListEmpty(Exception):
    pass

class sq:
    def __init__(self,list):
        self.l = list
        
    def shift(self):
        if(len(self.l)==0):
            raise ListEmpty("List is Empty")
        else:
            x=self.l[0]
            self.l.pop(0)
            return(x)
        
    def unshift(self,a):
        self.l.insert(0,a)
        print(self.l)
        
    def push(self,a):
        self.l.append(a)
        print(self.l)
        
    def pop(self):
        self.l.pop()
        print(self.l)
        
    def remove(self):
        x = max(self.l)
        self.l.remove(x)
        print(self.l)
        return(x)

try:
    x = sq([1,2,3,4])
    x.shift()
    x.unshift(9)
    x.push(99)
    x.pop()
    print(x.remove())
except Exception as e:
    print(e)
```

```python
class wordplay:
    def __init__(self,l):
        self.l = l
        
    def length(self,length):
        ans = []
        for i in self.l:
            if(len(i)==length):
                ans.append(i)
        return(ans)
    
    def start(self,start):
        ans = []
        for i in self.l:
            if(i.startswith(start)):
                ans.append(i)
        return(ans)
    
    def end(self,end):
        ans = []
        for i in self.l:
            if(i.endswith(end)):
                ans.append(i)
        return(ans)
    
    def pali(self):
        ans = []
        for i in self.l:
            if(i==i[::-1]):
                ans.append(i)
        return(ans)
    
    def only(self,only):
        ans = []
        for i in self.l:
            c = 0
            for j in only:
                if(j in i):
                    c+=1
            if(c==len(only)):
                ans.append(i)
        return(ans)
    
    def avoids(self,avoid):
        ans = []
        for i in self.l:
            c = 0
            for j in avoid:
                if(j in i):
                    c+=1
            if(c==0):
                ans.append(i)
        return(ans)
    
x =['apple','banana','find','dictionary','set','tuple','list','malayalam','nayan','grind','apricot']

y = wordplay(x)
print(y.length(5))
print(y.start('a'))
print(y.end('d'))
print(y.pali())
print(y.only('bna'))
print(y.avoids('amkd'))
```
